```python
import numpy as np
from sklearn.datasets import load_iris
import matplotlib.pyplot as plt
from sklearn.manifold import MDS
from sklearn.preprocessing import MinMaxScaler
from sklearn.decomposition import PCA
```

```python
data = load_iris()
X = data.data
```

```python
scaler = MinMaxScaler()
X_scaled = scaler.fit_transform(X)
```

```python
mds = MDS(2,random_state=0)
X_2d = mds.fit_transform(X_scaled)
```

```python
colors = ['red','green','blue']
plt.rcParams['figure.figsize'] = [7, 7]
plt.rc('font', size=14)
for i in np.unique(data.target):
  subset = X_2d[data.target == i]
  
  x = [row[0] for row in subset]
  y = [row[1] for row in subset]
  plt.scatter(x,y,c=colors[i],label=data.target_names[i])
plt.legend()
plt.show()
```

```python
pca = PCA(n_components=2)
x_pca = pca.fit_transform(X_scaled)
```

```python
colors = ['red','green','blue']
plt.rcParams['figure.figsize'] = [7, 7]
plt.rc('font', size=14)
for i in np.unique(data.target):
  subset = x_pca[data.target == i]
  
  x = [row[0] for row in subset]
  y = [row[1] for row in subset]
  plt.scatter(x,y,c=colors[i],label=data.target_names[i])
plt.legend()
plt.show()
```

```python
rows, columns = X_scaled.shape

columns_mean = np.average(X_scaled, axis=0)
col_means_row_vec = columns_mean.reshape((1, columns)) 
X_scaled_centered_np = X_scaled - col_means_row_vec

```

```python
u, s, vt = np.linalg.svd(X_scaled_centered_np.T, full_matrices=False)

K = 3 #K Number of singular components to select
k_pc_x = (np.diag(s[:K]) @ vt[:K])

colors = ['red','green','blue']
plt.rcParams['figure.figsize'] = [7, 7]
plt.rc('font', size=14)
for i in np.unique(data.target):
  subset = k_pc_x.T[data.target == i]
  
  x = [row[0] for row in subset]
  y = [row[1] for row in subset]
  plt.scatter(x,y,c=colors[i],label=data.target_names[i])
plt.legend()
plt.show()
```

```python
S = X_scaled @ X_scaled.T
```

```python
lamda, u  = np.linalg.eig(S)
```

```python
np.diag(np.sqrt(lamda[:2])).shape
```

```python
new_x = (np.diag(np.sqrt(lamda[:2])) @ u.T[:2])
new_x.shape
```

```python
colors = ['red','green','blue']
plt.rcParams['figure.figsize'] = [7, 7]
plt.rc('font', size=14)
for i in np.unique(data.target):
  subset = new_x.T[data.target == i]
  
  x = [row[0] for row in subset]
  y = [row[1] for row in subset]
  plt.scatter(x,y,c=colors[i],label=data.target_names[i])
plt.legend()
plt.show()
```
