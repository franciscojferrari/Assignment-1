```python
#Plotting fuction

def plot_scatter(data, title, ax = None, labels = True, fsize = (8,8), threeD = False):
    
    data.columns = data.columns.astype(str)
    
    if ax is  None:
        fig = plt.figure(figsize=fsize)
        ax = fig.add_subplot(111, projection = "3d" if threeD else None)
    
    # Get Unique types
    color_labels = data['type'].unique()
    # List of colors in the color palettes
    rgb_values = sns.color_palette("tab10", 7)
    # Map types to the colors
    color_map = dict(zip(color_labels, rgb_values))
    
    for color_label in color_labels:
        
        data_by_type = data[data['type'] == color_label]
        
        if threeD:
            ax.scatter(
                data_by_type["0"],
                data_by_type["1"],
                data_by_type["2"],
                c = data_by_type['type'].map(color_map),
                label=color_label)
            if labels:
                texts = [ ax.text(
                            x = data_by_type.iloc[i,2],
                            y = data_by_type.iloc[i,3],
                            z = data_by_type.iloc[i,4],
                            s = text, fontsize=7,
                            ha='center', va='center')
                        for i, text in enumerate(data_by_type['animal'])
                        ]
                adjust_text(texts)
        else:
            ax.scatter(
                data_by_type["0"],
                data_by_type["1"],
                c = data_by_type['type'].map(color_map),
                label=color_label)
            if labels:
                texts = [ ax.text(
                            x = data_by_type.iloc[i,2],
                            y = data_by_type.iloc[i,3],
                            s = text, fontsize=7,
                            ha='center', va='center')
                        for i, text in enumerate(data_by_type['animal'])
                        ]
                adjust_text(texts)
            
    
            
    ax.set_title(title, fontsize=15)
    ax.legend(loc=0, prop={'size': 6})
    if ax is  None:
        plt.show()
```

```python
def information_gain(data, target_name):
    '''
    Measures the reduction in entropy after the split  
    :param v: Pandas Series of the members
    :param split:
    :return:
    '''
    information_gain_df = pd.DataFrame()
    target = data[target_name]
    entropy_before = entropy( target.value_counts(normalize=True), base=2)
    
    for feature_name in data.drop([target_name], axis=1):
        
        feature = data[feature_name]
        
        grouped_distrib = target.groupby(feature) \
                            .value_counts(normalize=True) \
                            .reset_index(name='count') \
                            .pivot_table(index = feature_name, columns = target_name, values='count').fillna(0) 

        entropy_after = entropy(grouped_distrib, axis=1, base=2)
        entropy_after *= feature.value_counts(sort=False, normalize=True)
        
        information_gain = entropy_before - entropy_after.sum()
        
        information_gain_df = information_gain_df.append(
            {"feature":feature_name, "information_gain":information_gain}
            , ignore_index=True)
    return information_gain_df

a_h = animals_he.reset_index().drop('animal', axis=1)
information_gain = information_gain(a_h, 'type').sort_values(by=['information_gain'], ascending=False)
information_gain.head(10)
```

```python
animals_np = animals.to_numpy()
animals_he_np = animals_he.to_numpy()

rows, columns_o = animals_np.shape
rows, columns_he = animals_he_np.shape
```

#### Center the data

```python
columns_mean = np.average(animals_np, axis=0)
col_means_row_vec = columns_mean.reshape((1, columns_o)) 
animals_centered_np = animals_np - col_means_row_vec

animals_np = animals_np.T
animals_centered_np = animals_centered_np.T


columns_mean_he = np.average(animals_he_np, axis=0)
col_means_row_vec_he = columns_mean_he.reshape((1, columns_he)) 
animals_he_centered_np = animals_he_np - col_means_row_vec_he

animals_he_np = animals_he_np.T
animals_he_centered_np = animals_he_centered_np.T
```

#### Perform SVD

```python
#SVD Original dataset
u, s, vt = np.linalg.svd(animals_centered_np, full_matrices=False)

#SVD One-hot-encoded dataset
u_he, s_he, vt_he = np.linalg.svd(animals_he_centered_np, full_matrices=False)

K = 3 #K Number of singular components to select
animals_k_pca = (np.diag(s[:K]) @ vt[:K])
animals_he_k_pca = (np.diag(s_he[:K]) @ vt_he[:K])
```

## MDS

```python
from scipy.linalg import fractional_matrix_power
from scipy import linalg

def mds_from_original_data(Y, K):
    #Compute th similarity matrix (GramMatrix)
    S = Y.T @ Y
    # Compute the PC scores from Gram matrix
    lamda, vl, vr  = linalg.eig(S, left=True, right=True)
    # Double check the eigenvalues are positive.
    assert np.all(lamda[:K] > 0)
#     proj =fractional_matrix_power(np.diag(lamda[:K]) , 0.5) @ u.T[:K]
    proj = np.diag(np.sqrt(lamda[:K])) @ vl.T[:K]    
    return proj


def mds_from_distance_matrix(D, K, index = None, calculate_distance = False):
    
    if calculate_distance:
        D = pairwise_distances(D)
        
    centering_matrix = lambda n: np.identity(n) -(np.ones((n, 1)) @ np.ones((1, n)))/n
    
    N = D.shape[0]
    # Compute the Gram matrix from the distance_matrix
    gram_from_dist = -(centering_matrix(N) @ D @ centering_matrix(N))/2
    
    # Compute the PC scores from Gram matrix
    lamda, vl, vr  = linalg.eig(gram_from_dist, left=True, right=True)
    # Double check the eigenvalues are positive.
    assert np.all(lamda[:K] > 0)
#     proj =fractional_matrix_power(np.diag(lamda[:K]) , 0.5) @ u.T[:K]
    proj = np.diag(np.sqrt(lamda[:K])) @ vl.T[:K]
    
    return pd.DataFrame(data = proj.T.real, index=index).reset_index()
```
